```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

files.upload()

df = pd.read_csv('retail_sales_dataset.csv')

df.isnull().sum()

desc_stats = df.describe()

plt.figure(figsize=(15, 12))

plt.subplot(3, 2, 1)
sns.histplot(df['Total Amount'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Amount')

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.set_index('Date', inplace=True)
df_resampled = df['Total Amount'].resample('M').sum()

plt.subplot(3, 2, 2)
df_resampled.plot(color='orange')
plt.title('Sales Trend Over Time')
plt.ylabel('Total Sales')

plt.subplot(3, 2, 3)
sns.countplot(data=df, x='Gender', palette='viridis')
plt.title('Gender Distribution')

plt.subplot(3, 2, 4)
sns.barplot(x='Product Category', y='Total Amount', data=df, palette='plasma')
plt.xticks(rotation=90)
plt.title('Total Sales by Product Category')

plt.subplot(3, 2, 5)
df_monthly_sales = df.groupby(df.index.month)['Total Amount'].sum()
df_monthly_sales.plot(kind='line', marker='o', color='green')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)

plt.subplot(3, 2, 6)
sns.boxplot(x='Age', data=df)
plt.title('Age Distribution of Customers')

plt.subplots_adjust(wspace=0.3, hspace=0.4)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
sns.barplot(x='Product Category', y='Quantity', data=df, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Quantity Sold by Product Category')
plt.tight_layout()
plt.show()

print("Recommendations based on EDA: Focus on high-performing product categories for promotions.")
```
